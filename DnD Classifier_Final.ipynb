```python
#Methedology is based off a few key concepts for determining challenge of the bad guy
##  1) Input your group statistics as requested
##  2) AC Min and AC Max function as follows:
    #Easy Encounter: A 1 (Min) or a 5(Max) will hit the monster
    
    #Mid Encounter: A 6 (Min) or a 10(Max) will hit the monster
    
    #Hard Encounter: An 11 (Min) or a 15(Max) will hit the monster
    
    #Heroic Encounter: A 16 (Min) or a 20(Max) will hit the monster
    
##  3) HP is generated as follows:
    #Easy Encounter: Monster will last between 1 and 5 hits and can be hit by players rolling a 1 or greater at min, 
    #5 or greater for max  
    
    #Mid Encounter: Monster will last between 6 and 10 hits and can be hit by players rolling a 6 or greaterat min, 
    #10 or greater for max  
    
    #Hard Encounter: Monster will last between 11 and 15 hits and can be hit by players rolling an 11 or greater at min, 
    #15 or greater for max 
    
    #Heroic Encounter: Monster will last between 16 and 20 hits and can be hit by players rolling a 16 or greater at min, 
    #20 or greater for max 
    
##  4) To Hit is generated by keeping in mind the following:
    #Easy Encounter: Players can hit with a 1 or a 5, Monster can hit with a 16 through 20. Derived by taking AC-the needed
    #roll to hit, so for a Monster at Easy it would be player's AC-16 for Min To Hit. If AC=20, Monster gets a +4 Minimum
    
    #Mid Encounter: Players can hit with a 6 or a 10, Monster can hit with a 15 through 11. Derived by taking AC-the needed
    #roll to hit, so for a Monster at Mid it would be player's AC-11 for Min. If AC=20, Monster gets a +9 Minimum
    
    #Hard Encounter: Players can hit with a 11 or a 15, Monster can hit with a 10 through 6. Derived by taking AC-the needed
    #roll to hit, so for a Monster at Hard it would be player's AC-6 for Min. If AC=20, Monster gets a +14 Minimum
    
    #Heroic Encounter: Players can hit with a 16 or a 20, Monster can hit with a 1 through 5. Derived by taking AC-the 
    #needed roll to hit, so for a Monster at Heroic it would be player's AC for Min. If AC=20, Monster gets a +20 Minimum
    
##  5) Calculating the Damage Min and Max makes the following assumptions:
    ##Easy Encounter: Monster attacks can kill the average player with 5 hits at minimum and 1 hit at maximum
    
    ##Mid Encounter: Monster attacks can kill the average player with 10 hits at minimum and 6 hits at maximum
    
    ##Hard Encounter: Monster attacks can kill the average player with 15 hits at minimum and 11 hits at maximum
    
    ##Heroic Encounter: Monster attacks can kill the average player with 20 hits at minimum and 16 hits at maximum
```
